# 3.2 fine grained slection of words 
we can look at the long words of the text to see if they are more informative 

a.		{w | w  V & P(w)}

b.		[w for w in V if p(w)]


 This means "the set of all w such that w is an element of V (the vocabulary) and w has property P".

```python
V = set(text1)
long_words = [w for w in V if len(w) > 15]
sorted(long_words)
```
